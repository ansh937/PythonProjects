**4. Plot Facebook Views
python
Copy code
plt.bar([a-0.25 for a in days], facebook_views, color="y", label="facebook views")
[a-0.25 for a in days]: This list comprehension creates a new list where each day value is adjusted by subtracting 0.25. This shifts the bars for Facebook views slightly to the left.
plt.bar(..., facebook_views, color="y", label="facebook views"): This creates a bar plot with the adjusted days on the x-axis and facebook_views on the y-axis. The bars are colored yellow ("y") and labeled "facebook views".
5. Plot YouTube Views
python
Copy code
plt.bar([a+0.25 for a in days], youtube_views, color="b", label="youtube views")
[a+0.25 for a in days]: This list comprehension creates a new list where each day value is adjusted by adding 0.25. This shifts the bars for YouTube views slightly to the right.
plt.bar(..., youtube_views, color="b", label="youtube views"): This creates a bar plot with the adjusted days on the x-axis and youtube_views on the y-axis. The bars are colored blue ("b") and labeled "youtube views".
**

```python
facebook_views=[12,34,67,34,78,78,76,234,45,23,76,123,54,123,345,12,45,1,]
youtube_views=[23,34,56,8,89,12,45,23,65,67,23,45,67,266,24,56,67,23]
plt.figure(figsize=(10, 8))
days=range(1,len(facebook_views)+1)
plt.bar([a-0.25 for a in days],facebook_views,color="y",label="facebook views",width=0.25,align="edge")
plt.bar([a+0.25 for a in days],youtube_views,color="b",label="youtube views ",width=-0.25,align="edge")
plt.xlabel('Days')
plt.ylabel('Views')
plt.title('Daily views of the marketing channels')
plt.legend(loc="upper right")
plt.grid(True,linewidth=1,color="black",linestyle="--")
plt.show()
```
